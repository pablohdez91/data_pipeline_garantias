### Exporta a CSV

```python
def export_csv(df, file_path):
    csv_string = df.write_csv(separator=';', float_precision=10, float_scientific=False, datetime_format="%d/%m/%Y %H:%M:%S")
    try:
        with open(file_path, "w", encoding="cp1252", errors="replace") as f:
            f.write(csv_string)
        print(f"Archivo exportado exitosamente: {file_path}")
    except Exception as e:
        print(f"Error en la conversión para {file_path}: {e}")
```

```python
df_for_export = [
    (recupera_con_pagos_flujos, csv_recupera_con_pagos_flujos),
    (pagadas_global_vf, csv_pagadas_global_vf),
    (base_agrup_final, csv_base_agrup_final),
    (pagadas_agrup, csv_pagadas_agrup),
    (recup_agrup, csv_recup_agrup),
    (pagadas_detalle_vf, csv_pagadas_detalle_vf),
    (curva_recup, csv_curva_recup),
    (vf_foto_nr, csv_vf_foto_nr),
    (vf_foto_r, csv_vf_foto_r),
    (simples_microcredito, csv_simples_microcredito),
    (simples_empresarial, csv_simples_empresarial),
    (simples_resto, csv_simples_resto),
    (revolventes_microcredito, csv_revolventes_microcredito),
    (revolventes_empresarial, csv_revolventes_empresarial),
    (revolventes_resto, csv_revolventes_resto)
]
```

```python
for df, path in df_for_export:
    export_csv(df, path)
```
